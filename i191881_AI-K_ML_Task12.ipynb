```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Advertising.csv", index_col=0)
df
```

```python
# On the given dataset, implement gradient descent from scratch and find the linear relationship between features and output

# 1. Implement gradient descent from scratch
# 2. Find the linear relationship between features and output

# 1. Implement gradient descent from scratch
def gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)
    cost_history = np.zeros(iterations)
    theta_history = np.zeros((iterations, 2))
    for it in range(iterations):
        prediction = np.dot(X, theta)
        theta = theta - (1/m) * alpha * (X.T.dot((prediction - y)))
        theta_history[it, :] = theta.T
        cost_history[it]  = compute_cost(X, y, theta)
    return theta, cost_history, theta_history

# 2. Find the linear relationship between features and output
def compute_cost(X, y, theta):
    m = len(y)
    prediction = np.dot(X, theta)
    cost = (1/2*m) * np.sum(np.square(prediction-y))
    return cost

# 3. Plot the cost function
def plot_cost_function(cost_history, iterations):
    fig, ax = plt.subplots(figsize=(12,8))
    ax.set_ylabel('J(Theta)')
    ax.set_xlabel('Iterations')
    _ = ax.plot(range(iterations), cost_history, 'b.')
```

```python
# call the functions
X = np.c_[np.ones(df.shape[0]), df['TV']]
y = np.c_[df['sales']]
theta = np.random.randn(2,1)
iterations = 100

alpha = 0.00001
theta, cost_history, theta_history = gradient_descent(X, y, theta, alpha, iterations)
print("Theta: ",theta)
print("Cost History:",cost_history)
print("Theta History", theta_history)

```

```python
compute_cost(X, y, theta)
```

```python
plot_cost_function(cost_history, iterations)
    
```
