```python
df = pd.read_csv("/home/abdullah/Downloads/iris.csv")
df
```

```python
# Write a function that takes input from user and determines the input parameters belong which class (using Naïve assumptions)

def naive_bayes(x1,x2,x3,x4):

    mean1 = df[df['variety']=='Setosa'].mean()
    mean2 = df[df['variety']=='Versicolor'].mean()
    mean3 = df[df['variety']=='Virginica'].mean()

    std1 = df[df['variety']=='Setosa'].std()
    std2 = df[df['variety']=='Versicolor'].std()
    std3 = df[df['variety']=='Virginica'].std()

    p1 = stats.norm.pdf(x1,mean1[0],std1[0])*stats.norm.pdf(x2,mean1[1],std1[1])*stats.norm.pdf(x3,mean1[2],std1[2])*stats.norm.pdf(x4,mean1[3],std1[3])
    p2 = stats.norm.pdf(x1,mean2[0],std2[0])*stats.norm.pdf(x2,mean2[1],std2[1])*stats.norm.pdf(x3,mean2[2],std2[2])*stats.norm.pdf(x4,mean2[3],std2[3])
    p3 = stats.norm.pdf(x1,mean3[0],std3[0])*stats.norm.pdf(x2,mean3[1],std3[1])*stats.norm.pdf(x3,mean3[2],std3[2])*stats.norm.pdf(x4,mean3[3],std3[3])

    if p1>p2 and p1>p3:
        print("Class = Setosa")
    elif p2>p1 and p2>p3:
        print("Class = Versicolor")
    else:
        print("Class = Virginica")
        
naive_bayes(3.1,2.1,1.4,2.2)
    
```

```python
# Compare the above function with the inbuilt function of GaussianNB from sklearn

from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split

X = df.iloc[:,:-1]
y = df.iloc[:,-1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

gnb = GaussianNB()
gnb.fit(X_train,y_train)

y_pred = gnb.predict(X_test)

from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
        
```
